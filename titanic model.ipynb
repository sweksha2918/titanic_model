```python
def impuet_age(cols):
    age=cols[0]
    parch=cols[1]
    if pd.isnull(age):
        if parch==0:
            return 32
        elif parch==1:
            return 24
        elif parch==2:
            return 17
        elif parch==3:
            return 33
        elif parch==4:
            return 44
        else:
            return 43
    else:
        return age
```

```python
df['Age']=df[['Age','Parch']].apply(impuet_age,axis=1)
```

```python
df=pd.concat([df,sex,embarked],axis=1)
```

```python
df.drop(['Sex','Embarked'],axis=1,inplace=True)
```
